## **EDA and Preprocessing for Uber fair data**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('final_internship_data.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().mean()*100
```

```python
#Missing values represented less than 1% of the data, therefore dropping them had minimal impact on the available dat
df=df.dropna()
```

```python
df.describe()
```

```python
df.duplicated().sum()
```

```python
df=df.drop(columns=["User ID","User Name","Driver Name","pickup_datetime"])
```

```python
df.shape
```

## **Remove invalid rows / outliers**

```python
df = df[df['passenger_count'] > 0]
df = df[df['fare_amount'] > 0]
df = df[df['pickup_longitude'].between(-1.35, -1.25) & df['pickup_latitude'].between(0.68, 0.74)]
df = df[df['dropoff_longitude'].between(-1.35, -1.25) & df['dropoff_latitude'].between(0.68, 0.74)]
df = df[(df['distance'] > 0)]
```

```python
df.shape
```

```python
corr = df.corr(numeric_only=True)['fare_amount'].sort_values(ascending=False)
print(corr)
```

```python
# Convert categorical text columns to the 'category' dtype for memory efficiency
categorical_cols = ["Car Condition", "Weather", "Traffic Condition"]
df[categorical_cols] = df[categorical_cols].astype("category")
```

## **Split features and target**

```python
X=df.drop(['fare_amount'], axis=1)
Y=df['fare_amount']

# Identify numeric vs categorical feature columns for the pipeline
numeric_features = X.select_dtypes(include=["int64", "float64"]).columns.tolist()
categorical_features = X.select_dtypes(include=["category"]).columns.tolist()

print("Numeric features:", numeric_features)
print("Categorical features:", categorical_features)
```

## **Train and test split**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2, random_state=42)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.ensemble import RandomForestRegressor

# Numeric pipeline: fill missing values with the median because data has outliers, then scale
num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

# Categorical pipeline: fill missing values with the most frequent value bec data is categorical, then one-hot encode
cat_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Combine both transformers into a single preprocessing step
preprocessor = ColumnTransformer(transformers=[
    ("num", num_pipeline, numeric_features),
    ("cat", cat_pipeline, categorical_features)
])
```

## **Model Comparison**

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import r2_score, mean_squared_error,mean_absolute_error

models = {
    "Linear Regression": LinearRegression(),
    "Random Forest": RandomForestRegressor(random_state=42,n_jobs=-1),
    "Gradient Boosting": GradientBoostingRegressor(random_state=42)
}
results = []

for name, model in models.items():

    pipeline = Pipeline([
        ("preprocessor", preprocessor),
        ("selector", SelectKBest(score_func=f_regression, k=10)),
        ("model", model)
    ])

    pipeline.fit(X_train, Y_train)
    pred = pipeline.predict(X_test)

    r2 = r2_score(Y_test, pred)
    rmse = np.sqrt(mean_squared_error(Y_test, pred))
    mae=mean_absolute_error(Y_test, pred)

    results.append({
        "Model": name,
        "R²": r2,
        "RMSE": rmse,
        "MAE":mae
    })
    
```

```python
results_df = pd.DataFrame(results)
results_df
```

## **Build the final Pipeline**

```python
# Full pipeline: preprocessing -> feature selection -> model
# Select the 10 most relevant features based on their relationship with the target
model_pipeline = Pipeline(steps=[
    ("preprocessing", preprocessor),
    ("feature_selection", SelectKBest(score_func=f_regression, k=10)),
    ("model", RandomForestRegressor(random_state=42, n_jobs=-1))
])
```

## **Cross Validation**

```python
from sklearn.model_selection import cross_val_score

# 3-fold cross-validation on the training set using R2 as the scoring metric
cv_scores = cross_val_score(
    model_pipeline, X_train, Y_train, cv=3, scoring="r2", n_jobs=-1
)
#n_jobs=-1 → Use all available CPU cores, which usually makes cross-validation faster
print("Cross-validation R2 scores:", cv_scores)
print("Mean R2:", cv_scores.mean())
```

```python
model_pipeline.fit(X_train, Y_train)
Y_pred=model_pipeline.predict(X_test)
```

## **Evaluate the baseline pipeline**

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
mae=mean_absolute_error(Y_test,Y_pred)
mse=mean_squared_error(Y_test,Y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(Y_test,Y_pred)

print("Model Evaluation")
print(f"MAE  : {mae:.2f}")
print(f"MSE  : {mse:.2f}")
print(f"RMSE : {rmse:.2f}")
print(f"R²   : {r2:.4f}")
```

## **Hyperparameter tuning** 

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

param_dist = {
    "model__n_estimators": [100, 200],
    "model__max_depth": [15, 25, None],
    "model__min_samples_split": [2, 5],
    "model__min_samples_leaf": [1, 2],
}

random_search = RandomizedSearchCV(
    estimator=model_pipeline,
    param_distributions=param_dist,
    n_iter=8,          # Reduced from 20
    cv=3,              # Reduced from 5
    scoring="r2",
    random_state=42,
    n_jobs=1,
    verbose=2
)

random_search.fit(X_train, Y_train)

print("Best parameters found:", random_search.best_params_)
print("Best CV R2 score:", random_search.best_score_)
```

```python
# Retrieve the best pipeline (already trained on the full training data)
best_model = random_search.best_estimator_
best_model
```

## **Evaluate the tuned pipeline**

```python
best_pred = best_model.predict(X_test)
best_pred = best_model.predict(X_test)
best_mae = mean_absolute_error(Y_test, best_pred)
best_mse = mean_squared_error(Y_test, best_pred)
best_rmse = np.sqrt(best_mse)
best_r2 = r2_score(Y_test, best_pred)

print("Best Model Evaluation")
print(f"MAE  : {best_mae:.2f}")
print(f"MSE  : {best_mse:.2f}")
print(f"RMSE : {best_rmse:.2f}")
print(f"R²   : {best_r2:.4f}")
```

```python
comparison_df = pd.DataFrame({
    "Model": ["Baseline (Random Forest)", "Tuned (Random Forest)"],
    "MAE": [mae, best_mae],
    "RMSE": [rmse, best_rmse],
    "R2": [r2, best_r2]
})
comparison_df
```

```python
print("Original Model R² :", r2)
print("Tuned Model R²    :", best_r2)

if best_r2 > r2:
    print("The tuned model performed better.")
else:
    print("The original model performed better.")
```

## **Summary**

Random Forest was selected as the final model for deployment because it achieved the best performance on the test set. It produced the highest R² score (0.81), indicating that it explains the largest proportion of the variation in Uber fares, while also achieving the lowest RMSE and MAE, meaning its predictions are more accurate than the other models. Although Gradient Boosting performed similarly, Random Forest provided slightly better accuracy with fast prediction times, making it well suited for a real-time Flask application. Linear Regression is easier to interpret, but its prediction accuracy was significantly lower, so it was not chosen for deployment.
